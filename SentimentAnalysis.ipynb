```python
import csv
import pandas as pd
df = pd.read_csv(r"C:/Users/shali/OneDrive/Desktop/vs programs/III College Event Feedback Analysis/CleanedData/student_feedback.csv", encoding = "Windows-1252")

#NORMALIZE THE RATINGS TO (1-5) SCALE
rating_col = ["Well versed with the subject", "Explains concepts in an understandable way", "Use of presentations", "Degree of difficulty of assignments", "Solves doubts willingly", "Structuring of the course", "Provides support for students going above and beyond", "Course recommendation based on relevance"]
df[rating_col] = ((df[rating_col]-1)) / (10-1) * (5-1) + 1
df[rating_col] = df[rating_col].round(2)
df.to_csv(r"C:/Users/shali/OneDrive/Desktop/vs programs/III College Event Feedback Analysis/DataToWorkOn/student_feedback1.csv", index=False)
# print(df[rating_col].min().min(), df[rating_col].max().max())

# SENTIMENT ANALYSIS
df["Avg_rating"] = df[rating_col].mean(axis=1)

def sentiment(avg):
    if avg >= 4:
        return "Positive"
    elif avg >= 3:
        return "Negative"
    else:
        return "Neutral"
df["Overall Sentiment"] = df["Avg_rating"].apply(sentiment)
df.to_csv(
    r"C:/Users/shali/OneDrive/Desktop/vs programs/III College Event Feedback Analysis/DataToWorkOn/student_feedback_with_sentiment.csv",
    index=False
)

# CONSOLIDATED RATINGS
df = pd.read_csv(r"C:\Users\shali\OneDrive\Desktop\vs programs\III College Event Feedback Analysis\DataToWorkOn\student_feedback_with_sentiment.csv")
rating_columns = [
    "Well versed with the subject",
    "Explains concepts in an understandable way",
    "Use of presentations",
    "Degree of difficulty of assignments",
    "Solves doubts willingly",
    "Structuring of the course",
    "Provides support for students going above and beyond",
    "Course recommendation based on relevance"
]
df_long = df.melt(
    value_vars=rating_columns,
    var_name="Parameter",
    value_name="Rating"
)
df_long.to_csv(r"C:\Users\shali\OneDrive\Desktop\vs programs\III College Event Feedback Analysis\DataToWorkOn\ratings_long.csv", index=False)
```
